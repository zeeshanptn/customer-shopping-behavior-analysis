## 4. Data Cleaning

### Handling Missing Values

```python
# Imputing missing values in Review Rating column with the median rating of the product category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

### Renaming Columns

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

### Creating Age Groups

```python
# create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

### Creating Purchase Frequency (Days)

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

### Checking Redundant Columns

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

### Removing Redundant Columns

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```

## 5. Exploratory Data Analysis (EDA)

### 5.1 Distribution of Purchase Amount

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df["purchase_amount"], bins=20, kde=True)

plt.title("Distribution of Purchase Amount")
plt.xlabel("Purchase Amount")
plt.ylabel("Frequency")

plt.show()
```

### 5.2 Total Sales by Product Category

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x="category",
    y="purchase_amount",
    estimator="sum"
)

plt.title("Total Sales by Category")
plt.xticks(rotation=45)
plt.show()
```

### 5.3 Average Purchase Amount by Gender

```python
plt.figure(figsize=(6,5))

sns.barplot(
    data=df,
    x="gender",
    y="purchase_amount",
    estimator="mean"
)

plt.title("Average Purchase by Gender")
plt.show()
```

### 5.4 Customer Age Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(df["age"], bins=15)

plt.title("Age Distribution")
plt.show()
```

### 5.5 Average Purchase Amount by Age Group

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x="age_group",
    y="purchase_amount",
    estimator="mean"
)

plt.title("Average Purchase by Age Group")
plt.show()
```

### 5.6 Total Sales by Season

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x="season",
    y="purchase_amount",
    estimator="sum"
)

plt.title("Sales by Season")
plt.show()
```

### 5.7 Subscription Status vs Average Purchase Amount

```python
plt.figure(figsize=(6,5))

sns.barplot(
    data=df,
    x="subscription_status",
    y="purchase_amount",
    estimator="mean"
)

plt.title("Subscription vs Purchase Amount")
plt.show()
```

### 5.8 Sales by Payment Method

### 5.10 Correlation Analysis

```python
plt.figure(figsize=(6,4))

numeric_df = df.select_dtypes(include="number")

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")
plt.show()
```

### 5.11 Top 10 Locations by Sales

```python
top_states = (
    df.groupby("location")["purchase_amount"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_states.index,
    y=top_states.values
)

plt.xticks(rotation=45)
plt.title("Top 10 Locations by Sales")
plt.show()
```
